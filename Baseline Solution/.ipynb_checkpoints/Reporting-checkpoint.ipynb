```python
import os

raw_data_path = r"Dataset\samples"
# Access all subfolders
subfolders = [x[0] for x in os.walk(raw_data_path) if x[0]!=raw_data_path]
print(subfolders)
```

```python
import json
import pandas as pd

#i=1
df=pd.DataFrame()
for s in subfolders:
    student = s.split('\\')[-1]
    jfiles = [x for x in os.listdir(s) if x.endswith('.json')]
    if len(jfiles)==1:
        with open(os.path.join(s, jfiles[0]), 'r', encoding='utf-8', errors='ignore') as jf:
            j = json.load(jf)
            try:
                data=[j['nativeProficiency'],j['englishProficiency']]
           # print(data)
            except:
                data=[j['nativeProficiency'],'None']
            temp=pd.DataFrame([data],columns=['Native Reading Level','English Level'])
            #print(temp)
            df=df.append(temp)
```

```python
import pandas as pd
import matplotlib.pyplot as plt        # For plotting graphs 
%matplotlib inline 
import warnings                        # To ignore any warnings warnings.filterwarnings("ignore")
from matplotlib import style

style.use('ggplot')

dataset = pd.DataFrame(df, 
               columns =['Native Reading Level', 'English Level'])
#print(dataset.head())
shape=dataset.shape
plot1=plt.figure(1) 
plt.subplot(221) 
dataset['Native Reading Level'].value_counts().plot.bar(figsize=(20,10), title= 'Native Proficiency') 
plt.subplot(222) 
#dataset['mathematicsProficiency'].value_counts().plot.bar(figsize=(20,10),title= 'Mathematics Proficiency') 
#plt.subplot(223) 
dataset['English Level'].value_counts().plot.bar(figsize=(20,10),title= 'English Proficiency') 
plt.suptitle('Total Samples: '+str(shape[0]),fontsize=25) 
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
